## **The Movie Scenery: A Tour Through Ratings and Preferences**

Our first step is to get acquainted with the dataset. We have at our disposal two key datasets: one containing movie information and the other with user ratings. Let's begin by understanding the structure of our data.

The movies dataset consists of a wealth of information, such as movie IDs, titles, and genres. We have a diverse range of movies spanning various genres, from action-packed adventures to heartwarming romances.

The ratings dataset provides us with a closer look at how users feel about these movies. It reveals the collective sentiments of movie enthusiasts, and we are eager to delve into this treasure trove of opinions.

## **Unveiling Movie Based preferences**

### **Examine the most popular movies based on ratings and genres**

```python
popularity_df = df_ratings.groupby('movieId')['rating'].count().reset_index()
popularity_df.rename(columns={'rating': 'total_ratings'}, inplace=True)
popularity_df = popularity_df.merge(df_movies[['movieId', 'title', 'genres']], on='movieId', how='left')
popularity_df = popularity_df.sort_values(by='total_ratings', ascending=False)
```

```python
# Set the number of top movies to display
top_n_movies = 10

# Sort the dataframe in descending order based on total_ratings
top_movies_df = popularity_df.sort_values(by='total_ratings', ascending=False).head(top_n_movies)

# Create a bar plot to visualize the top-rated movies
plt.figure(figsize=(12, 6))
sns.barplot(x='total_ratings', y='title', data=top_movies_df, palette='viridis')
plt.title(f"Top {top_n_movies} Movies by Total Ratings")
plt.xlabel("Total Ratings")
plt.ylabel("Movie Title")
plt.show()
```

```python
# Create a bar plot to visualize the top-rated movies with their genres
plt.figure(figsize=(12, 6))
sns.barplot(x='total_ratings', y='genres', data=top_movies_df, palette='viridis')
plt.title(f"Top {top_n_movies} Movies by Total Ratings and Genres")
plt.xlabel("Total Ratings")
plt.ylabel("Genres")
plt.show()
```

### **Examine movies with relatively fewer ratings but high average ratings**

```python
# Group by movieId and calculate total ratings and average rating
hidden_gems_df = df_ratings.groupby('movieId')['rating'].agg(['count', 'mean']).reset_index()
hidden_gems_df.rename(columns={'count': 'total_ratings', 'mean': 'average_rating'}, inplace=True)

# Merge with 'df_movies' to get movie information
hidden_gems_df = hidden_gems_df.merge(df_movies[['movieId', 'title', 'genres']], on='movieId', how='left')

# Set a threshold for minimum ratings (You can change this value based on your preference)
min_ratings_threshold = 50
hidden_gems_df = hidden_gems_df[hidden_gems_df['total_ratings'] >= min_ratings_threshold]

# Sort the dataframe in descending order based on average rating
hidden_gems_df = hidden_gems_df.sort_values(by='average_rating', ascending=False)
```

```python
# Set the number of top hidden gems to display
top_n_hidden_gems = 10

# Sort the dataframe in descending order based on average_rating
top_hidden_gems_df = hidden_gems_df.sort_values(by='average_rating', ascending=False).head(top_n_hidden_gems)

# Create a bar plot to visualize the top hidden gems with their average ratings
plt.figure(figsize=(12, 6))
sns.barplot(x='average_rating', y='title', data=top_hidden_gems_df, palette='viridis')
plt.title(f"Top {top_n_hidden_gems} Hidden Gems by Average Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Movie Title")
plt.show()
```

**"Hidden gems" refer to movies that are exceptionally good or valuable but may not have received as much attention or recognition as the blockbuster or widely popular movies. They are those underrated or less-promoted movies that often surprise and delight viewers with their exceptional content. They are simply those movies waiting to be discovered by a wider audience**

### **Compare Popularity and Hidden Gems**

```python
# Combine popularity_df and hidden_gems_df for box plot
combined_df = pd.concat([popularity_df.assign(is_popular='Popular'), hidden_gems_df.assign(is_popular='Hidden Gem')])

# Create a box plot to visualize the distribution of ratings between popular movies and hidden gems
plt.figure(figsize=(8, 6))
sns.boxplot(x='is_popular', y='total_ratings', data=combined_df, palette='viridis')
plt.title("Distribution of Ratings: Popular Movies vs. Hidden Gems")
plt.xlabel("Category")
plt.ylabel("Total Ratings")
plt.show()
```

```python
from surprise import Reader, Dataset, NMF, KNNBasic, NormalPredictor, BaselineOnly
# Using SVD from the surprise library as an example
svd_model = SVD()

# Initialize NMF model
nmf_model = NMF()

# Initialize BaselineOnly model
baseline_model = BaselineOnly()

# Initialize KNNBasic model
knn_model = KNNBasic()

# Initialize NormalPredictor model
normal_model = NormalPredictor()
```

```python
# Merge with imdb_data to get "director," "runtime," and "plot keywords"
train = pd.merge(train, imdb_data[['movieId', 'director']], on='movieId', how='left')

# Merge with imdb_data to get "director," "runtime," and "plot keywords"
test = pd.merge(test, imdb_data[['movieId', 'director']], on='movieId', how='left')
```

```python
train.head()
```

```python
train_data, val_data = train_test_split(train, test_size=0.2, random_state=42)
```

```python
# Create a surprise dataset from the training data
from surprise import Dataset
from surprise import Reader

reader = Reader(rating_scale=(0.5, 5))
train_dataset = Dataset.load_from_df(train_data[['userId', 'movieId', 'rating']], reader)
trainset = train_dataset.build_full_trainset()

# Train the model on the training set
svd_model.fit(trainset)
nmf_model.fit(trainset)
baseline_model.fit(trainset)
knn_model.fit(trainset)
normal_model.fit(trainset)
```

```python
# Create a surprise dataset from the validation data
val_dataset = Dataset.load_from_df(val_data[['userId', 'movieId', 'rating']], reader)
valset = val_dataset.construct_testset(val_dataset.raw_ratings)

# Predict ratings for the validation set
val_predictions_svd = svd_model.test(valset)

# Predict ratings for the validation set
val_predictions_nmf = nmf_model.test(valset)

# Predict ratings for the validation set
val_predictions_baseline = baseline_model.test(valset)

# Predict ratings for the validation set
val_predictions_knn = knn_model.test(valset)

# Predict ratings for the validation set
val_predictions_normal = normal_model.test(valset)

# Calculate RMSE for evaluation
val_rmse_svd = np.sqrt(mean_squared_error([pred.r_ui for pred in val_predictions_svd], [pred.est for pred in val_predictions_svd]))

# Calculate RMSE for evaluation
val_rmse_nmf = np.sqrt(mean_squared_error([pred.r_ui for pred in val_predictions_nmf], [pred.est for pred in val_predictions_nmf]))

# Calculate RMSE for evaluation
val_rmse_baseline = np.sqrt(mean_squared_error([pred.r_ui for pred in val_predictions_baseline], [pred.est for pred in val_predictions_baseline]))

# Calculate RMSE for evaluation
val_rmse_knn = np.sqrt(mean_squared_error([pred.r_ui for pred in val_predictions_knn], [pred.est for pred in val_predictions_knn]))

# Calculate RMSE for evaluation
val_rmse_normal = np.sqrt(mean_squared_error([pred.r_ui for pred in val_predictions_normal], [pred.est for pred in val_predictions_normal]))
```
